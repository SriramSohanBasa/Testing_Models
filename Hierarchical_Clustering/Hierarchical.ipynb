```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
%matplotlib inline
```

```python
iris = datasets.load_iris()
```

```python
iris
```

```python
print(iris.DESCR)
```

```python
print(iris.feature_names)
```

```python
iris_data=pd.DataFrame(iris.data)
```

```python
iris_data.columns=iris.feature_names
```

```python
iris_data.head()
```

```python
print(iris.target_names)
```

```python
print(iris.target)
```

```python
#0 represents setosa, 1 represents versicolor, 2 represents virginica
```

```python
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
```

```python
X_scaled= scaler.fit_transform(iris_data)
```

```python
X_scaled
```

```python
##apply PCA
from sklearn.decomposition import PCA
```

```python
pca= PCA(n_components=2)
```

```python
pca_scaled = pca.fit_transform(X_scaled)
```

```python
plt.scatter(pca_scaled[:,0], pca_scaled[:,1], c=iris.target)
```

```python
##agglomerative clustering, construct a dendogram
import scipy.cluster.hierarchy as sch
```

```python
##plotting dendogram
plt.figure(figsize=(10,7))
plt.title("Dendogram")

##create dendogram
sch.dendrogram(sch.linkage(pca_scaled, method='ward'))
plt.title("Dendogram")
plt.xlabel("sample index")
plt.ylabel("eucledian distance")
plt.show()
```

```python
##find longest vertical line without crossing any horizontal line
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=2, linkage='ward')
cluster.fit(pca_scaled)
```

```python
cluster.labels_ ##this shows the cluster to which each sample belongs to
```

```python
plt.scatter(pca_scaled[:,0], pca_scaled[:,1], c=cluster.labels_)
```
